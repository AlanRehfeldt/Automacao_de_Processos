### Passo 1 - Importar Arquivos e Bibliotecas

```python
# Importar bibliotecas
import pandas as pd
import pathlib
import yagmail
```

```python
#Importar bases de dados
emails = pd.read_excel(r"Bases de dados/Emails.xlsx")
lojas = pd.read_csv(r"Bases de dados/Lojas.csv", encoding="latin1", sep = ";")
vendas = pd.read_excel(r"Bases de dados/Vendas.xlsx")
display(emails)
display(lojas)
display(vendas)
```

### Passo 2 - Criar uma Tabela para cada Loja e Definir o dia do Indicador

```python
# Incluir nome da loja na tabela de vendas
vendas = vendas.merge(lojas, on = "ID Loja")
display(vendas)
```

```python
# Criar uma tabela para cada loja
dicionario_lojas = {}

for loja in lojas["Loja"]:
    dicionario_lojas[loja] = vendas.loc[vendas["Loja"]==loja, :]
    
display(dicionario_lojas["Rio Mar Recife"])
display(dicionario_lojas["Shopping Vila Velha"])
```

```python
# Identificando a data do arquivo
dia_indicador = vendas["Data"].max()
print(dia_indicador)
print("{}/{}".format(dia_indicador.day, dia_indicador.month))
```

### Passo 3 - Salvar a planilha na pasta de backup

```python
# Identificar se a pasta ja existe
caminho_backup = pathlib.Path(r"Backup Arquivos Lojas")
arquivos_pasta_backup = caminho_backup.iterdir()

lista_nomes_backup = []

for arquivo in arquivos_pasta_backup:
    lista_nomes_backup.append(arquivo.name)

print(lista_nomes_backup)

for loja in dicionario_lojas:
    if loja not in lista_nomes_backup:
        nova_pasta = caminho_backup / loja
        nova_pasta.mkdir()
        
    # Salvar dentro da pasta    
    nome_arquivo = "{}_{}_{}.xlsx".format(dia_indicador.month, dia_indicador.day, loja)
    print(nome_arquivo)
    local_arquivo = caminho_backup / loja / nome_arquivo
    dicionario_lojas[loja].to_excel(local_arquivo)
```

### Passo 4 - Calcular indicadores de todas as lojas e enviar E-mail para cada gerente

```python
# Definição de metas

meta_faturamento_dia = 1000
meta_faturamento_ano = 1650000
meta_qntd_produtos_dia = 4
meta_qntd_produtos_ano = 120
meta_ticket_medio_dia = 500
meta_ticket_medio_ano = 500
```

```python
usuario = yagmail.SMTP(user = "inserir e-mail do remetente", password = "Inserir senha da conta do e-mail")

for loja in dicionario_lojas:
    vendas_loja = dicionario_lojas[loja]
    vendas_loja_dia = vendas_loja.loc[vendas_loja["Data"] == dia_indicador, :]

    # Faturamento
    faturamento_ano = vendas_loja["Valor Final"].sum()
    faturamento_dia = vendas_loja_dia["Valor Final"].sum()

    # Diversidade produtos
    qntd_produtos_ano = len(vendas_loja["Produto"].unique())
    qntd_produtos_dia = len(vendas_loja_dia["Produto"].unique())

    # Ticket médio
    valor_venda = vendas_loja.groupby("Código Venda").sum()
    ticket_medio_ano = valor_venda["Valor Final"].mean()
    valor_venda_dia = vendas_loja_dia.groupby("Código Venda").sum()
    ticket_medio_dia = valor_venda_dia["Valor Final"].mean()
    
    
    gerente = emails.loc[emails["Loja"]==loja, "Gerente"].values[0]
      
    corpo_email = f'''
    <p>Bom dia {gerente},</p> 
    <p>O resultado de ontem <strong>({dia_indicador.day}/{dia_indicador.month})</strong> da loja <strong>{loja}</strong> foi.</p>
    
    <table>
      <tr>
        <th>Indicador</th>
        <th>Valor dia</th>
        <th>Meta dia</th>
      </tr>
      <tr>
        <td>Faturamento</td>
        <td style="text-align: center">R${faturamento_dia:.2f}</td>
        <td style="text-align: center">R${meta_faturamento_dia:.2f}</td>
      </tr>
      <tr>
        <td>Diversidade de produtos</td>
        <td style="text-align: center">{qntd_produtos_dia}</td>
        <td style="text-align: center">{meta_qntd_produtos_dia}</td>
      </tr>
      <tr>
        <td>Ticket médio</td>
        <td style="text-align: center">R${ticket_medio_dia:.2f}</td>
        <td style="text-align: center">R${meta_ticket_medio_dia:.2f}</td>
      </tr>
    </table>
    <br>
    <table>
      <tr>
        <th>Indicador</th>
        <th>Valor ano</th>
        <th>Meta ano</th>
      </tr>
      <tr>
        <td>Faturamento</td>
        <td style="text-align: center">R${faturamento_ano:.2f}</td>
        <td style="text-align: center">R${meta_faturamento_ano:.2f}</td>
      </tr>
      <tr>
        <td>Diversidade de produtos</td>
        <td style="text-align: center">{qntd_produtos_ano}</td>
        <td style="text-align: center">{meta_qntd_produtos_ano}</td>
      </tr>
      <tr>
        <td>Ticket médio</td>
        <td style="text-align: center">R${ticket_medio_ano:.2f}</td>
        <td style="text-align: center">R${meta_ticket_medio_ano:.2f}</td>
      </tr>
    </table>
    <br>
    
    <p>Segue em anexo a planilha com todos os dados para mais detalhes.</p>
    <p>Qualquer dúvida estou a disposição.</p>
    
    <p>Att.,</p>
    <p>Alan Rehfeldt</p>
    '''
    anexo = pathlib.Path.cwd() / caminho_backup / loja / "{}_{}_{}.xlsx".format(dia_indicador.month, dia_indicador.day, loja)
    
    usuario.send(to = emails.loc[emails["Loja"]==loja, "E-mail"].values[0], subject = f"OnePage do dia {dia_indicador.day}/{dia_indicador.month} - Loja {loja}", contents = corpo_email, attachments = str(anexo))
```

### Passo 5 - Criar ranking para diretoria

```python
faturamento_loja = vendas.groupby("Loja")[["Loja", "Valor Final"]].sum()
faturamento_loja_ano = faturamento_loja.sort_values(by = "Valor Final", ascending = False)
display(faturamento_loja_ano)

nome_arquivo = '{}_{}_Ranking Anual.xlsx'.format(dia_indicador.month, dia_indicador.day)
local_arquivo = caminho_backup / nome_arquivo
faturamento_loja_ano.to_excel(local_arquivo)

vendas_dia = vendas.loc[vendas['Data']==dia_indicador, :]
faturamento_lojas_dia = vendas_dia.groupby('Loja')[['Loja', 'Valor Final']].sum()
faturamento_lojas_dia = faturamento_lojas_dia.sort_values(by='Valor Final', ascending=False)
display(faturamento_lojas_dia)

nome_arquivo = '{}_{}_Ranking Dia.xlsx'.format(dia_indicador.month, dia_indicador.day)
local_arquivo = caminho_backup / nome_arquivo
faturamento_lojas_dia.to_excel(local_arquivo)
```

```python
gerente = emails.loc[emails["Loja"]==loja, "Gerente"].values[0]

corpo_email = f'''
Presados, bom dia,

Melhor loja do dia em faturamento {faturamento_lojas_dia.index[0]} com faturamento R${faturamento_lojas_dia.iloc[0, 0]:.2f}
Pior loja do dia em faturamento {faturamento_lojas_dia.index[-1]} com faturamento R${faturamento_lojas_dia.iloc[-1, 0]:.2f}

Melhor loja do ano em faturamento {faturamento_loja_ano.index[0]} com faturamento R${faturamento_loja_ano.iloc[0, 0]:.2f}
Pior loja do ano em faturamento {faturamento_loja_ano.index[-1]} com faturamento R${faturamento_loja_ano.iloc[-1, 0]:.2f}


Segue em anexo os rankings do ano e dos dias de todas as lojas.

Qualquer dúvida estou a disposição.

Att.
Alan
'''
anexo1 = str(pathlib.Path.cwd() / caminho_backup / "{}_{}_Ranking Anual.xlsx".format(dia_indicador.month, dia_indicador.day))
anexo2 = str(pathlib.Path.cwd() / caminho_backup / "{}_{}_Ranking Dia.xlsx".format(dia_indicador.month, dia_indicador.day))

usuario.send(to = emails.loc[emails["Loja"]=="Diretoria", "E-mail"].values[0], subject = "Ranking do dia {}/{}".format(dia_indicador.day, dia_indicador.month), contents = corpo_email, attachments = [anexo1, anexo2])
```
